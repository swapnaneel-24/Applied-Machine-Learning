## What is NumPy?
NumPy (Numerical Python) is an open-source Python library used for numerical computations and working with large multidimensional arrays and matrices.

### It provides:
    1. Fast array operations
    2. Mathematical functions
    3. Linear algebra tools
    4. Random number generation
    5. Statistical functions
### NumPy forms the foundation of many libraries such as:
    1. Pandas
    2. Scikit-Learn
    3. TensorFlow
    4. PyTorch
    5. SciPy
    6. OpenCV

## Why NumPy was **Developed**?

#### Python lists were:
    1. Slow
    2. Memory inefficient
    3. Not suitable for large numerical computations
### NumPy was developed to:
    1. Perform high-speed mathematical calculations
    2. Store large datasets efficiently
    3. Support matrix operations
    4. Provide vectorized computations

- NumPy arrays facilitate advanced mathematical and other types of operations on large numbers of data. Typically, such operations are executed more efficiently and with less code than is possible using Python’s built-in sequences.

- A growing plethora of scientific and mathematical Python-based packages are using NumPy arrays; though these typically support Python-sequence input, they convert such input to NumPy arrays prior to processing, and they often output NumPy arrays.

## NumPy Array Copy vs View


### **Copy vs View in NumPy**

1. **Copy creates a new array**, while **View refers to the original array**.
2. **Copy has its own data**, so changes in the copy do **not** affect the original array.
3. **Changes in the original array do not affect the copy**.
4. **View shares the data with the original array**.
5. **Changes in the view affect the original array**.
6. **Changes in the original array also affect the view**.

### Shortcut to Remember

* **Copy → Independent data → No effect on original**
* **View → Shared data → Changes affect both**


```python
a = np.array([2,3,5])
b = a.copy()
```

```python
a , b
```

```python
b[0] = 1111134567
```

```python
a , b
```

```python
a = np.array([2,3,5])
b = a.view()
a,b
```

```python
a[1] = 4567
```

```python
a,b
```

```python
b[0] = 56789765456
```

```python
a,b
```

### **Checking Data Ownership in NumPy**

1. Every NumPy array has a **`base` attribute**.
2. If **`base` returns `None`**, the array **owns its data** (it is a **copy**).
3. If **`base` returns another array**, the array **does not own its data** (it is a **view**).
4. A **copy is independent**, while a **view shares data with the original array**.



```python
a = np.array([2,3,5])
b_view = a.view()
c_copy = a.copy()

print(a)
print(b_view)
print(c_copy)
```

```python
print(a.base)
print(b_view.base)
print(c_copy.base)
```

### **Iterating Arrays Using `nditer()`**

1. **`nditer()` is a function used for easy array iteration.**
2. **It can handle both simple and advanced iterations.**
3. **`nditer()` accesses each scalar element one by one.**
4. **It simplifies iteration for multi-dimensional arrays.**
5. **Using `nditer()`, we do not need multiple nested `for` loops.**

### Note:

* **`nditer()` → Easy iteration**
* **Visits each element one by one**
* **Useful for high-dimensional arrays**
* **Avoids writing many nested loops**


```python
a = np.array([[23,45,66,5,6,6],[44,56,77,8,9,4]])
```

```python
for val in np.nditer(a):
    print(val)
```

```python
a = np.array([[[23,45,66],[5,6,6]],[[44,56,77],[8,9,4]]])
for val in np.nditer(a):
    print(val)
```

### **Iterating Arrays with Different Data Types**

1. **`op_dtypes` is used to specify the desired data type while iterating.**
2. **NumPy does not change the original array elements directly.**
3. **It uses extra temporary memory called a buffer to store converted values.**
4. **To enable this buffer, use `flags=['buffered']` with `nditer()`.**
5. **This allows iteration with a different data type without modifying the original array.**

### Short Note

* **`op_dtypes` → Change data type during iteration**
* **Original array remains unchanged**
* **Buffer = Temporary memory**
* **Use `flags=['buffered']` to enable the buffer**


```python
a = np.array([2,4,5])
a
```

```python
for val in np.nditer(a, flags=['buffered'], op_dtypes=['S']):
    print(val)
```

### **Enumerated Iteration Using `ndenumerate()`**

1. **Enumeration means accessing elements along with their index.**
2. **`ndenumerate()` is used to iterate through arrays with index values.**
3. **It returns both the index and the corresponding element.**
4. **It is useful when we need the position of each element during iteration.**
5. **It works with arrays of any dimension.**



```python
a = np.array([[2,3,4],[5,6,7]])
for index, val in np.ndenumerate(a):
    print(f"index : {index} and value: {val}")
```

## NumPy Joining Array

### **Joining Arrays in NumPy**

1. **Joining means combining two or more arrays into a single array.**
2. **NumPy joins arrays based on axes.**
3. **The `concatenate()` function is used to join arrays.**
4. **Multiple arrays are passed as a sequence to `concatenate()`.**
5. **The `axis` parameter specifies along which axis the arrays are joined.**
6. **If `axis` is not specified, NumPy uses `axis=0` by default.**

### Remember

* **Joining = Combining arrays**
* **Use `concatenate()`**
* **Join is done along an axis**
* **Default axis = 0**


```python
a1 = np.array([1,2,3])
a2 = np.array([11,21,33])
arr = np.concatenate((a2,a1))
arr
```

```python
arr = np.concatenate((a2,a1), axis = 0)
arr
```

### **Joining Arrays Using Stack Functions**

1. **Stacking is similar to concatenation, but it creates a new axis.**
2. **The `stack()` function is used to stack arrays.**
3. **Stacking combines arrays one above the other or side by side along a new dimension.**
4. **Multiple arrays are passed as a sequence to `stack()`.**
5. **The `axis` parameter specifies the new axis along which arrays are stacked.**
6. **If `axis` is not specified, NumPy uses `axis=0` by default.**



* **Stacking = Concatenation + New Axis**

### **Splitting NumPy Arrays**

1. **Splitting is the opposite of joining.**
2. **Joining combines multiple arrays into one array.**
3. **Splitting divides one array into multiple arrays.**
4. **The `array_split()` function is used to split arrays.**
5. **Pass the array and the number of splits to `array_split()`.**

### Notesz

* **Joining → Many arrays → One array**
* **Splitting → One array → Many arrays**
* **Use `array_split()` for splitting**


```python
arr = np.array([1,2,4,5,6,7,8,9])
splitted_array = np.array_split(arr,3)
```
